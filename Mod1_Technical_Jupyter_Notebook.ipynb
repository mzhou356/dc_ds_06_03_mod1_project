## Genre and seasonalities: 
  * box plot for all 6 genres by month 

```python
# load the genre_popularity.pkl produced from Movie Project-Detailed Data Processing.py
genre_popularity = pd.read_pickle('data/genre_popularity.pkl')
# set popularity to be in scale log to better handle large outliers 
genre_popularity.popularity = np.log(genre_popularity.popularity)
```

```python
# create genre count dataframe from genre_popularity 
genre_count = genre_popularity.groupby('genres').size().reset_index()
# create genre dict from genre_count
genre_dict = {}
for i, row in genre_count.iterrows():
    genre_dict[row['genres']]= row[0]
# create genre_dict_prob 
# total counts
total = sum(genre_dict.values())
for i,v in genre_dict.items():
    genre_dict[i] = v/total*100
```

```python
# turn genre_dict into a dataframe genre_prop dataframe 
genre_prop = pd.DataFrame.from_dict(genre_dict, orient = 'index').reset_index()
# rename column using rename_columns function from hf 
hf.rename_columns(genre_prop, {'index': 'genres', 0: 'proportion'})
# sort the genre by proportion and we will do a pie chart for top 10 most prevalant genres 
genre_prop_sorted = genre_prop.sort_values('proportion', ascending = False)
fig1, ax1 = plt.subplots()
ax1.pie(genre_prop_sorted.proportion[:10], labels=genre_prop_sorted.genres[:10], autopct='%1.1f%%', startangle=90)
plt.show() 
```

* chosen the top 6 most frequent genres: Drama, comedy, action, horror, documentary, thriller:

```python
# genre_prop data frame
genre_prop.head()
```

```python
# plot top 6 most common genres by month 
plt.figure(figsize=(20, 10))
for i, genre in enumerate(['Drama','Comedy','Action','Horror','Documentary','Thriller'],1):
    ax = plt.subplot(2,3,i)
    df = genre_popularity[genre_popularity.genres == genre]
    sns.boxplot(x = 'month', y = 'popularity', data = df, ax=ax)
    ax.set_xlabel('Month')
    ax.set_ylabel('Popularity')
    ax.set_title(genre+' Popularity by Month')
```

```python
# look at popularity among month and genres separately 
# month only 
ax = sns.boxplot(x = 'month', y = 'popularity', data = genre_popularity)
plt.title("Popularity of Movies by Release Month")
```

```python
# genre only
ax = sns.boxplot(x = 'genres', y = 'popularity', data = genre_popularity)
plt.title("Popularity of Movies by Release Genre")
for item in ax.get_xticklabels():
    item.set_rotation(90)
```

```python
# second popularity dataframe for studying genres and month 
genre_popularity.head()  
```

## Hiring known actors and actress:
  * is more the better? 

```python
# load the dataset produced from Movie_Project_Detailed_Data_Processing.py
performer_popularity = pd.read_pickle('data/perform_popularity.pkl')
# set popularity to be in scale log 
performer_popularity.popularity = np.log(performer_popularity.popularity)
# plot group by num_known_actor_actress
plt.figure(figsize=(10, 8))
bplot = sns.boxplot(y='popularity', x='num_known_actor_actress',
                    data=performer_popularity,
                    width=0.5,
                    palette="colorblind")
plt.title('popularbity distribution for num of known performers', fontdict = {'fontsize':16})
bplot.set_xlabel("num_known_actors_actresses",fontsize=15)
bplot.set_ylabel("popularity(Log)",fontsize=15)
plt.hlines(y=1.1, xmin=0.5, xmax=5.5, linestyle='--', color='red')
plt.show()
```

### hiring 4 known actor_actress may be better than hiring too many or too few 

## Movie Length: 
  * divide the length into 4 bins:
    * 0 to 50 min
    * 50 min to 100 min
    * 100 min to 150 min
    * 150 min to 200 min

```python
# use binning helper function from hf to bin the run time into 4 categories
performer_popularity.runtime_minutes = performer_popularity.runtime_minutes.apply(
    hf.binning)
# plot group by runtime_minutes
plt.figure(figsize=(8, 6))
bplot = sns.boxplot(y='popularity', x='runtime_minutes',
                    data=performer_popularity,
                    width=0.5,
                    palette="colorblind", order=["under 50 min", "50 to 100 min", "100 to 150 min", 'over 150 min'])
plt.title('popularity distribution for different runtimes',
          fontdict={'fontsize': 16})
bplot.set_xlabel("movie duration (min)", fontsize=15)
bplot.set_ylabel("popularity(Log)", fontsize=15)
plt.hlines(y=1.75, xmin=1, xmax=3, linestyle='--', color='red')
plt.show()
```

### 100 to 150 min seems to be the best runtime length

## Examine the budget vs worldwide gross revenue
 * spend big make big or spend small and make big margin 
  



```python
# convert production_budget and worldwide_gross in log scale 
budget_popularity.production_budget = np.log(budget_popularity.production_budget)
budget_popularity.worldwide_gross = np.log(budget_popularity.worldwide_gross)
```

```python
# set the budget threshold to be at least log(budget)> 10 (e^10)
revenue_by_budget = budget_popularity.loc[budget_popularity.production_budget > 10, ['worldwide_gross', 'production_budget']]
# bin log budget from 11 to 20 
revenue_by_budget.production_budget = revenue_by_budget.production_budget.apply(lambda x: int(x))
plt.figure(figsize=(8, 6))
bplot = sns.boxplot(y='worldwide_gross', x='production_budget',
                    data=revenue_by_budget,
                    width=0.5,
                    palette="colorblind")
plt.title('Log Worldwide Revenue by Budget Threshold',
          fontdict={'fontsize': 16})
bplot.set_xlabel('Threshold Log Production Budget', fontsize=15)
bplot.set_ylabel('Log Worldwide Sales', fontsize=15)

plt.show()
```
